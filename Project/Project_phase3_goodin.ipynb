```python

import tkinter as tk
from tkinter import messagebox
from tkinter import ttk
import mysql.connector

#########################################################################################################################

def open_facility_details_gui():

    def find_facility_details(state, population_type, service_type):
        try:
            # Connect to MySQL server
            cnx = mysql.connector.connect(user='root', password='Desi20!^', host='127.0.0.1', database='project_phase3_goodiin')
            cursor = cnx.cursor()
    
            # Execute SQL query with placeholders
            query = ("SELECT F.CName, F.St_Address, F.Suite, F.City, F.Zip, F.Phone "
                     "FROM FACILITY F "
                     "JOIN FACILITY_LOCATIONS FL ON F.FNumber = FL.FNumber "
                     "JOIN PROGRAM P ON F.FNumber = P.FNumber "
                     "JOIN POPULATION POP ON P.PID = POP.PID "
                     "JOIN SERVICE S ON P.PID = S.PID "
                     "WHERE FL.State = %s AND POP.PType = %s AND S.SType = %s")
            cursor.execute(query, (state, population_type, service_type))
    
            # Fetch results
            results = cursor.fetchall()
    
            # Close cursor and connection
            cursor.close()
            cnx.close()
    
            return results
    
        except mysql.connector.Error as err:
            print(f"Error: {err}")
            return []

    def search_facility_details():
        state = state_entry.get()
        population_type = population_type_entry.get()
        service_type = service_type_entry.get()
        
        if not state or not population_type or not service_type:
            messagebox.showerror("Error", "Please enter all fields")
            return
        
        facility_details = find_facility_details(state, population_type, service_type)
        if not facility_details:
            messagebox.showinfo("No Facilities", "No facilities found for the given criteria")
        else:
            # Clear previous results
            for row in result_treeview.get_children():
                result_treeview.delete(row)
            
            # Insert new results
            for details in facility_details:
                result_treeview.insert("", "end", values=details)
    
    # Create main window
    root = tk.Tk()
    root.title("Facility Details Search")
    
    # Create input fields
    state_label = tk.Label(root, text="State:")
    state_label.grid(row=0, column=0, sticky="e")
    state_entry = tk.Entry(root)
    state_entry.grid(row=0, column=1)
    
    population_type_label = tk.Label(root, text="Population Type:")
    population_type_label.grid(row=1, column=0, sticky="e")
    population_type_entry = tk.Entry(root)
    population_type_entry.grid(row=1, column=1)
    
    service_type_label = tk.Label(root, text="Service Type:")
    service_type_label.grid(row=2, column=0, sticky="e")
    service_type_entry = tk.Entry(root)
    service_type_entry.grid(row=2, column=1)
    
    search_button = tk.Button(root, text="Search", command=search_facility_details)
    search_button.grid(row=3, column=0, columnspan=2)
    
    # Result table
    columns = ("Facility Name", "Street Address", "Suite", "City", "Zip", "Phone")
    result_treeview = ttk.Treeview(root, columns=columns, show="headings")
    
    for col in columns:
        result_treeview.heading(col, text=col)
    
    result_treeview.grid(row=4, column=0, columnspan=2)
    
    root.mainloop()
    
    pass

####################################################################################################

def open_programs_in_city_gui():

    def find_programs_in_city(state, city):
        try:
            # Connect to MySQL server
            cnx = mysql.connector.connect(user='root', password='Desi20!^', host='127.0.0.1', database='project_phase3_goodiin')
            cursor = cnx.cursor()
    
            # Execute SQL query with placeholders
            query = ("SELECT DISTINCT F.CName, P.PID, P.PName, S.SType, POP.PType "
                     "FROM SERVICE S "
                     "JOIN PROGRAM P ON S.PID = P.PID "
                     "JOIN FACILITY F ON P.FNumber = F.FNumber "
                     "JOIN FACILITY_LOCATIONS FL ON F.FNumber = FL.FNumber "
                     "JOIN POPULATION POP ON P.PID = POP.PID "
                     "WHERE F.City = %s AND FL.State = %s")
            cursor.execute(query, (city, state))
    
            # Fetch results
            results = cursor.fetchall()
    
            # Close cursor and connection
            cursor.close()
            cnx.close()
    
            return results
    
        except mysql.connector.Error as err:
            print(f"Error: {err}")
            return []
    
    def search_programs():
        state = state_entry.get()
        city = city_entry.get()
        
        if not state or not city:
            messagebox.showerror("Error", "Please enter all fields")
            return
        
        programs = find_programs_in_city(state, city)
        if not programs:
            messagebox.showinfo("No Programs", "No programs found for the given criteria")
        else:
            # Clear previous results
            for row in result_treeview.get_children():
                result_treeview.delete(row)
            
            # Insert new results
            for program in programs:
                result_treeview.insert("", "end", values=program)
    
    # Create main window
    root = tk.Tk()
    root.title("Program Search")
    
    # Create input fields
    state_label = tk.Label(root, text="State:")
    state_label.grid(row=0, column=0, sticky="e")
    state_entry = tk.Entry(root)
    state_entry.grid(row=0, column=1)
    
    city_label = tk.Label(root, text="City:")
    city_label.grid(row=1, column=0, sticky="e")
    city_entry = tk.Entry(root)
    city_entry.grid(row=1, column=1)
    
    search_button = tk.Button(root, text="Search", command=search_programs)
    search_button.grid(row=2, column=0, columnspan=2)
    
    # Result table
    columns = ("Facility Name", "PID", "Program Name", "Service Type", "Population Type")
    result_treeview = ttk.Treeview(root, columns=columns, show="headings")
    
    for col in columns:
        result_treeview.heading(col, text=col)
    
    result_treeview.grid(row=3, column=0, columnspan=2)
    
    root.mainloop()
    pass

###############################################################################################

def open_reserve_spot_gui():
    def find_programs_in_city(city):
        try:
            # Connect to MySQL server
            cnx = mysql.connector.connect(user='root', password='Desi20!^', host='127.0.0.1', database='project_phase3_goodiin')
            cursor = cnx.cursor()
    
            # Execute SQL query with placeholders
            query = ("SELECT P.PID, P.PName, S.SType, POP.PType, S.Available_Spots "
                     "FROM PROGRAM P "
                     "JOIN FACILITY F ON P.FNumber = F.FNumber "
                     "JOIN FACILITY_LOCATIONS FL ON F.FNumber = FL.FNumber "
                     "JOIN SERVICE S ON P.PID = S.PID "
                     "JOIN POPULATION POP ON P.PID = POP.PID "
                     "WHERE F.City = %s")
            cursor.execute(query, (city,))
    
            # Fetch results
            results = cursor.fetchall()
    
            # Close cursor and connection
            cursor.close()
            cnx.close()
    
            return results
    
        except mysql.connector.Error as err:
            print(f"Error: {err}")
            return []
    
    def reserve_spot(city, program_choice, num_reserved):
        try:
            programs = find_programs_in_city(city)
            selected_program = programs[program_choice - 1]
            pid, pname, stype, ptype, available_spots = selected_program
    
            if num_reserved > available_spots:
                messagebox.showerror("Error", "Not enough available spots.")
                return
    
            # Connect to MySQL server
            cnx = mysql.connector.connect(user='root', password='Desi20!^', host='127.0.0.1', database='project_phase3_goodiin')
            cursor = cnx.cursor()
    
            # Update available spots
            update_query = ("UPDATE SERVICE "
                            "SET Available_Spots = %s "
                            "WHERE PID = %s")
            cursor.execute(update_query, (available_spots - num_reserved, pid))
            cnx.commit()
    
            messagebox.showinfo("Success", f"You have reserved {num_reserved} spot(s) for Program: {pname}, Service Type: {stype}, Population Type: {ptype}.\nAvailable spots left: {available_spots - num_reserved}")
    
            # Close cursor and connection
            cursor.close()
            cnx.close()
    
        except ValueError:
            messagebox.showerror("Error", "Invalid input. Please enter a valid number.")
        except IndexError:
            messagebox.showerror("Error", "Invalid choice. Please enter a valid program number.")
        except mysql.connector.Error as err:
            messagebox.showerror("Error", f"Database error: {err}")
    
    def display_programs():
        city = city_entry.get()
        programs = find_programs_in_city(city)
        if not programs:
            messagebox.showinfo("No Programs", "No programs found for the given city.")
        else:
            # Clear previous results
            for row in result_treeview.get_children():
                result_treeview.delete(row)
    
            # Insert new results
            for idx, program in enumerate(programs, start=1):
                result_treeview.insert("", "end", values=(idx,) + program[1:])
    
    # Create main window
    root = tk.Tk()
    root.title("Program Reservation")
    
    # Create input fields
    city_label = tk.Label(root, text="City:")
    city_label.grid(row=0, column=0, sticky="e")
    city_entry = tk.Entry(root)
    city_entry.grid(row=0, column=1)
    
    search_button = tk.Button(root, text="Search Programs", command=display_programs)
    search_button.grid(row=0, column=2)
    
    program_choice_label = tk.Label(root, text="Program Choice:")
    program_choice_label.grid(row=1, column=0, sticky="e")
    program_choice_entry = tk.Entry(root)
    program_choice_entry.grid(row=1, column=1)
    
    num_reserved_label = tk.Label(root, text="Number of Spots to Reserve:")
    num_reserved_label.grid(row=2, column=0, sticky="e")
    num_reserved_entry = tk.Entry(root)
    num_reserved_entry.grid(row=2, column=1)
    
    reserve_button = tk.Button(root, text="Reserve", command=lambda: reserve_spot(city_entry.get(), int(program_choice_entry.get()), int(num_reserved_entry.get())))
    reserve_button.grid(row=3, column=0, columnspan=2)
    
    # Result table
    columns = ("Program Choice", "Program Name", "Service Type", "Population Type", "Available Spots")
    result_treeview = ttk.Treeview(root, columns=columns, show="headings")
    
    for col in columns:
        result_treeview.heading(col, text=col)
    
    result_treeview.grid(row=4, column=0, columnspan=3, padx=10, pady=10)
    
    root.mainloop()
    
    
    pass

############################################################################################################

# Define function to open selected GUI based on user's selection
def open_selected_gui(selection):
    if selection == "Find Facility Details":
        open_facility_details_gui()
    elif selection == "Find Programs in City":
        open_programs_in_city_gui()
    elif selection == "Reserve Spot":
        open_reserve_spot_gui()
    else:
        messagebox.showerror("Error", "Invalid selection")

# Define function to create the selection screen
def create_selection_screen():
    # Create the main window
    root = tk.Tk()
    root.title("Function Selection")

    # Create a label
    label = tk.Label(root, text="Select a function:")
    label.pack(pady=10)

    # Define function options
    functions = ["Find Facility Details", "Find Programs in City", "Reserve Spot"]

    # Create buttons for each function
    for func_name in functions:
        button = tk.Button(root, text=func_name, command=lambda func=func_name: open_selected_gui(func))
        button.pack()

    root.mainloop()

# Call the function to create the selection screen
create_selection_screen()
```
